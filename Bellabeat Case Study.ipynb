## Merging df & df1

#### Grouping the users into active levels

#### Defining function filter

```python
def filter (x):
    if x <= 7499:
        return 'sedentary'
    if (x > 7499 and x <= 10499):
        return 'active'
    if x > 10499:
        return 'very active'
    
```

#### Applying the filter function to totalsteps column

```python
merged_df['activelevel'] = merged_df['totalsteps'].apply(filter)

merged_df
```

> since day0fweek_x & day0fweek_y are the same, and weekday_x & weekday_y are the same I'm going to drop 2 of those 4 columns and rename them for ease of use. 

#### Dropping unwanted columns

## Insights

### Calories v Steps

```python
sns.scatterplot(x = 'totalsteps', y = 'calories',  data = merged_df, hue = 'activelevel')
plt.title('Correlation between Calories & Total Steps')
plt.show()
```

> This graph shows the sedentary users don't walk/run as much as the active and very active users but burn around the same amount of calories on average.

### Average number of steps per day

```python
day_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday' 'Saturday', 'Sunday']
fig, ax = plt.subplots(1,1, figsize = (9, 6))

day_grp = merged_df.groupby(['day_of_week'])
avg_daily_steps = day_grp['totalsteps'].mean()
avg_steps = merged_df['totalsteps'].mean()

plt.bar(avg_daily_steps.index, avg_daily_steps)

ax.set_xticks(range(len(day_of_week)))
ax.set_xticklabels(day_of_week)

ax.axhline(y = avg_daily_steps.mean(), color = 'green', label = 'Avg daily steps')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Number of steps')
ax.set_title('Avg number of steps per day')

plt.legend()
plt.show()
```

> On Tuesday, Wednesday & Saturday the number of steps taken surpass the average, every other day the average is beneath the average 

### Percentage of activity in minutes

```python
sedentaryminutes = merged_df['sedentaryminutes'].sum()

lightlyactiveminutes = merged_df['lightlyactiveminutes'].sum()

fairlyactiveminutes = merged_df['fairlyactiveminutes'].sum()

veryactiveminutes = merged_df['veryactiveminutes'].sum()


slices = [sedentaryminutes, lightlyactiveminutes, fairlyactiveminutes, veryactiveminutes]
labels = ['sedentary minutes', 'lightlyactiveminutes', 'fairlyactiveminutes', 'veryactiveminutes']
explode = [0, 0, 0, 0.1]
plt.pie(slices, labels = labels, explode = explode, autopct = '%1.1f%%', textprops = dict(size = 9), shadow = False)

plt.title('Percentage of activity in minutes', fontsize = 18)
plt.tight_layout()

plt.show()
```

> This chart shows that the users spend most of their time sedentary and the lowest amount of time fairly active, this may suggest there's a short period of time between being sedentary and being very active or vice versa.
